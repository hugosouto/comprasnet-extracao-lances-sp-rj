```python
import csv

file = 'dados/tbl_lances.csv'

lines = []
with open(file,'r') as f:
    for line in f.readlines():
        lines.append(line[:-1])

with open(file,'w') as correct:
    writer = csv.writer(correct, dialect = 'excel')
    with open('file', 'r') as mycsv:
        reader = csv.reader(mycsv)
        for row in reader:
            if row[0] not in lines:
                writer.writerow(row)

if '\0' in open('file').read():
    print("Null bytes detected in the input file.")
else:
    print("Null bytes not detected in the input file.")

if '\b' in open('file').read():
    print("Line break detected in the input file.")
else:
    print("Line break not detected in the input file.")
```
